```python
def preprocess_text(text):
    # Sentence tokenization using Indic NLP
    sentences = sentence_split(text, lang="ta")
    processed_sentences = []

    for sentence in sentences:
        words = list(indic_tokenize.trivial_tokenize(sentence))
        words = [word for word in words if word not in stop_words]
        processed_sentences.append(" ".join(words))

    return processed_sentences
```

```python
def compute_tf_score(sentences):
  if not sentences:
        return [0] * len(sentences)
  tfidf = TfidfVectorizer()
  tfidf_matrix = tfidf.fit_transform(sentences).toarray()
  return np.mean(tfidf_matrix, axis=1)
```

```python
def compute_position_score(sentences):
    position_scores = []
    for i, _ in enumerate(sentences):
        position_scores.append(1 - (i / len(sentences)))
    return position_scores
```

```python
def compute_length_score(sentences):
    max_length = max(len(sentence) for sentence in sentences)
    return [len(sentence) / max_length for sentence in sentences]
```

```python
def extract_summary(text, threshold=0.1):
    try:

      sentences = preprocess_text(text)
      tf_scores = compute_tf_score(sentences)
      position_scores = compute_position_score(sentences)
      length_scores = compute_length_score(sentences)


      final_scores = [
          0.7 * tf + 0.1 * pos + 0.1 * length
          for tf, pos, length in zip(tf_scores, position_scores, length_scores)
      ]


      summary = [sentence for score, sentence in zip(final_scores, sentences) if score >= threshold]
      return " ".join(summary)
    except:
      return " "
```

```python
from rouge import Rouge

rouge = Rouge()

def calculate_rouge_scores(reference_summary, generated_summary):
  """
  Calculate ROUGE scores between a reference summary and a generated summary.
  """
  scores = rouge.get_scores(generated_summary, reference_summary)
  rouge_1_f1 = scores[0]['rouge-1']['f']
  rouge_2_f1 = scores[0]['rouge-2']['f']
  rouge_l_f1 = scores[0]['rouge-l']['f']
  return {
      'rouge-1': rouge_1_f1,
      'rouge-2': rouge_2_f1,
      'rouge-l': rouge_l_f1
  }


# Example usage:
reference_summary = sample_text  # Replace with your reference summary
generated_summary = summary  # Replace with your generated summary

rouge_scores = calculate_rouge_scores(reference_summary, generated_summary)
print("ROUGE Scores:", rouge_scores)
```
